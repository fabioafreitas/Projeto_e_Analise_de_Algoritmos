```python
def activity_selector(s, f):
  j = 0
  act = [0]
  for i in range(len(f)):
    if s[i] >= f[j]:
      act.append(i)
      j = i
  return act
```

```python
s = [1, 3, 0, 5, 3, 5, 6, 8, 8, 2, 12]
f = [4, 5, 6, 7, 9, 9, 10, 11, 12, 14, 16]

print('atividades selecionadas (nos Ã­ncices do cormen):')
[i+1 for i in activity_selector(s,f)]
```

# 2Âº)

Considere uma modificaÃ§Ã£o no problema de seleÃ§Ã£o de atividades jÃ¡ vistos. Agora, cada atividade $a_i$ tem, em adiÃ§Ã£o ao seu tempo de inÃ­cio e tempo de tÃ©rmino, um valor monetÃ¡rio de $v_i$. O objetivo agora nÃ£o Ã© mais maximizar o nÃºmero de atividades agendadas, mas sim maximizar o valor monetÃ¡rio das atividades agendadas. Ou seja, deseja-se encontrar um conjunto A de atividades compativeis tal que o somatÃ³ro abaixo seja maximizado. DÃª um algoritmo com custo polinomial no tempo.

\begin{align*}
\sum_{a_k \in A}^{}v_k
\end{align*}

resposta:

```python
# Dado um vetor de atividades de tamanho i busco o 
# indice da primeira atividade que se encaixa com a 
# Ãºltima atividade desse vetor se nÃ£o houver uma, 
# retorno -1, se sim retoro seu indice
def latest_non_conflict(s, f, i):
  for j in reversed(range(i)):
    if f[j] <= s[i-1]:
      return j
  return -1



def activity_selector_max_profit_aux(s, f, v, n, m):
  # caso base
  if n == 1: 
    return v[n-1]
  # memoization - verifica se subproblema ja foi resolvido
  elif m[n-1] > float('-inf'):
    return m[n-1]
  # caso geral do problema
  else:
    # verificando valor final com o valor da atividade n-1 
    profit1 = v[n-1]
    i = latest_non_conflict(s, f, n)
    if i != -1:
      profit1 = profit1 + activity_selector_max_profit_aux(s, f, v, i+1, m)
    
    # verificando valor final sem o valor da atividade n-1 
    profit2 = activity_selector_max_profit_aux(s, f, v, n-1, m)

    # adiciona maior caso no memoization e o retorna
    m[n-1] = max(profit1, profit2)
    return m[n-1]




def activity_selector_max_profit(start, finish, value):
  num_activities = len(start)
  memoization = np.full(num_activities, float('-inf'))
  memoization[0] = value[0]
  return activity_selector_max_profit_aux(start, finish, value, num_activities, memoization)
```

```python
# assume-se que as atividades estao ordenadas pelo tempo de finalizaÃ§Ã£o
start =  [1, 3, 0, 5, 3, 5, 6, 8, 8, 2, 12]
finish = [4, 5, 6, 7, 9, 9, 10, 11, 12, 14, 16]
value =  [4, 5, 5, 3, 5, 6, 6, 7, 9, 9, 0]

print('max: ')
print(activity_selector_max_profit(start, finish, value))
```

# 3Âº)

Suponha que lhe sÃ£o dados dois conjuntos $A$ e $B$, cada qual contendo $n$ inteiros positivos. VocÃª pode escolher reordenar os conjuntos de qualquer forma. ApÃ³s o reordenamento, seja $a_i$ o i-Ã©simo elemento do conjunto $A$, e seja $b_i$ o i-Ã©simo elemento do conjunto $B$. VocÃª irÃ¡ receber um pagamento dado ela produtÃ³rio abaixo. DÃª um algorimto que maximiza o seu pagamento e mostre qual Ã© o seu custo em tempo.

\begin{align*}
\prod_{i=1}^{n}a_i^{b_i}
\end{align*}

resposta:

Para maximizar o pagamento, devemos calcular a exponenciaÃ§Ã£o com ambos os arrays ordenados de corma crescente, de modo que sempre teremos as maiores bases elevadas aos maiores expoentes. Para o caso de existir $0$ na lista $A$, entÃ£o o custo do algoritmo serÃ¡ $\Omega(n)$. Caso contrÃ¡rio, devemos ordenar os dois arrays e em seguida realizar o cÃ¡lculo do produtÃ³rio. Esse custo pode ser calculado somando os custos do caso $else$, descrito no algoritmo abaixo.

\begin{align*}
O(n*log(n) + n*log(n) + n) = O(n*log(n))
\end{align*}

Desta forma, o custo final serÃ¡ de:

$$
\mathrm{Custo} = \begin{cases}
    \Omega(n) \text{, se 0} \in A\\
    O(n*log(n))
\end{cases}
$$

```python
def calcular_salario(A, B):             
  if 0 in A:                            # n           custo de buscar o 0 num array de tamanho n
    return 0                            # 1           constante
  else:                                 # 1           constante
    A = np.sort(A, kind='mergesort')    # n*log(n)    custo do mergesort Ã© n*log(n)
    B = np.sort(B, kind='mergesort')    # n*log(n)    custo do mergesort Ã© n*log(n)
    total = 1                           # 1           constante           
    for i in range(len(A)):             # n           loop com as n iteraÃ§Ãµes do produtÃ³rio
      total = total * A[i]**B[i]        # 1           constante
    return total                        # 1           constante
```

```python
num = 6
A = np.random.randint(low=1,high=num+1, size=num) # lista de nÃºmeros aleatorios entre 1 e 4
B = np.random.randint(low=1,high=num+1, size=num) # lista de nÃºmeros aleatorios entre 1 e 4
print("A = {}\nB = {}\nTotal = {}".format(A,B,calcular_salario(A,B)))
```
